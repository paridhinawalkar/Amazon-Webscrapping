```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
Product_name = []
price =[]
desc = []
star = []
rate =[]
date = []
```

```python
for j in range(2,21) :
    
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.5735.199 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    }
    
    url = "https://www.amazon.in/s?k=mobile+under+50000+5g+phone&page=2&crid=1L5PO4R427AG9&qid=1732019983&sprefix=%2Caps%2C432&ref=sr_pg_"+str(j)

    response = requests.get(url, headers=headers)

    print(response.status_code)
    print(j)

    B = BeautifulSoup(response.text,"lxml")

    soup = B.find("div",{"class" : "s-main-slot s-result-list s-search-results sg-row"})

    #Product
    Product = soup.find_all("span",{"class" : "a-size-medium a-color-base a-text-normal"})

    print(len(Product))

    for i in Product :
        name = i.text.split("(")[0].strip()
        #print(name)
        Product_name.append(name)

    #Price
    Price = soup.find_all("span",{"class" : "a-price-whole"})

    print(len(Price))

    for i in Price :
        price.append(i.text)
        #print(i.text)

    #Description
    Desc = soup.find_all("span" , {"class" : "a-size-medium a-color-base a-text-normal"})

    print(len(Desc))
    for i in Desc :
        if "(" in i.text:
            d = i.text.split("(")[1].strip(")")
        else:
            d = "NaN"  
        desc.append(d)
        #print(d)

    #Stars
    Star = soup.find_all("span",{"class" : "a-icon-alt"})

    print(len(Star))
    for i in Star :
        s = i.text.split(" ")[0]
        star.append(s)
        #print(s)

    #Ratings
    Rating = soup.find_all("span",{"class" : "a-size-base s-underline-text"})

    print(len(Rating))

    for i in Rating :
        #print(i.text)
        rate.append(i.text)

    time.sleep(2)
```

```python
dictionary = {"Product" : Product_name , "Price" : price , "Description" : desc , "Stars" : star , "Ratings" : rate }
```

```python
max_length = max(len(v) for v in dictionary.values())

for k, v in dictionary.items():
    dictionary[k] += [None] * (max_length - len(v))  

df = pd.DataFrame(dictionary)
```
